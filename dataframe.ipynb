```python
import pandas as pd
import numpy as np
from csv_tools.csv_changer import change_csv

data = pd.read_csv("csv_tools/imdb_top_1000.csv")

data_filtered = data.drop(["Poster_Link", "Certificate", "Overview", "Star1", "Star2", "Star3", "Star4", "Gross"], axis=1)
```

```python
# Clean Dataframe
data_clean = data_filtered.head(50)

data_clean
```

```python
# Dirty Dataframe
data_dirty = data_clean.copy()

# Putting Nan values
nan_mask = np.random.rand(*data_dirty.shape)<0.4
data_dirty[nan_mask] = np.nan

# Changing data types
type_mask_1 = np.random.rand(len(data_dirty))<0.5
type_mask_2 = np.random.rand(len(data_dirty))<0.9
columns = ["Meta_score", "IMDB_Rating", "No_of_Votes", "Released_Year"]
# Changing for an empty string 
data_dirty.loc[type_mask_1, columns] = ""
# Changing to a string number
data_dirty.loc[type_mask_2, columns] = data_dirty.loc[type_mask_2, columns].astype(str)

data_dirty
```

```python
data_dirty.to_csv("csv_tools/data.csv", index=False)
data_clean.to_csv("csv_tools/data_clean.csv", index=False)
```
